# groupby

Agrupar por Name y mostrar el conteo de los datos de las demás columnas


```python
df_pokemon.groupby('Name').count()
```

Agrupar por Name y mostrar la media de los datos de las demás columnas


```python
df_pokemon.groupby('Name').mean()
```

La columna NAME, en los casos anteriores, pasa a ser el índice.

Podemos usar loc y acceder a un dato específico del DataFrame. Agrupar por autor y mostrar la suma de los valores de las demás columnas

```python
df_pokemon.groupby('Name').max()
```

```python
df_pokemon.groupby('Name').min()
```

```python
df_pokemon.groupby('Name').sum()
```

La función agg() permite aplicar varias funciones al DataFrame una vez agrupado según una columna específica. Agrupar por Author y mostrar el mínimo y máximo de las demás columnas


```python
df_pokemon.groupby('Name').agg(['min','max'])
```

```python
df_pokemon.groupby('Name').agg({'Type 2':['min','max'], 'Legendary':'sum'})
```
